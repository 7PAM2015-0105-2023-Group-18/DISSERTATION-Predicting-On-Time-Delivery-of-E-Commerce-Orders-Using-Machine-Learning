# Data Analysis

```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px

# Set style
sns.set_style('darkgrid')

# Function to analyze distribution of categorical variable
def visualize_categorical(feature, chart_title):
    print(f"Total distinct values in {chart_title}: {len(feature.value_counts())}\n")

    value_counts = feature.value_counts()
    
    print(value_counts)
    num_categories = len(value_counts)

    # Creating bar chart
    plt.figure(figsize=(16, 6))
    plt.subplot(1, 2, 1)
    plt.title(f'{chart_title} - Bar Chart', fontsize=16)
    plt.xticks(rotation=45)
    bar_plot = sns.barplot(x=value_counts.index, y=value_counts.values, edgecolor="white", palette=sns.color_palette("icefire"))
    
    total_count = len(feature)
    for bar in bar_plot.patches:
        percent = '{:.1f}%'.format(100 * bar.get_height() / total_count)
        x_position = bar.get_x() + bar.get_width() / 2 - 0.08
        y_position = bar.get_y() + bar.get_height()
        bar_plot.annotate(percent, (x_position, y_position), size=12)

    # Creating pie chart
    plt.subplot(1, 2, 2)
    plt.title(f'{chart_title} - Pie Chart', fontsize=16)
    labels = value_counts.index
    plt.pie(x=value_counts, autopct="%.1f%%", explode=[0.02] * num_categories, labels=labels, pctdistance=0.5)
    plt.show()


# Function to plot distributions for numerical features
def plot_numerical(feature, title):
    plt.figure(figsize=(10, 6))
    plt.title(f'{title} - Distribution', fontsize=16)
    sns.histplot(feature, kde=True, bins=30)
    plt.show()
```

```python
import numpy as np

# Function to analyze relationship of categorical feature with target
def analyze_categorical_relationship(feature, target, data, target_outcome=1):
    colormaps = ['viridis', 'plasma', 'inferno', 'magma', 'cividis',
                 'coolwarm', 'afmhot', 'gist_heat', 'hot', 'copper',
                 'jet', 'rainbow', 'seismic', 'twilight', 'twilight_shifted']
    selected_colormap = colormaps[np.random.randint(1, len(colormaps))]

    category_to_ratio = {}
    for category in data[feature].unique():
        outcome_count = data[(data[target] == target_outcome) & (data[feature] == category)].shape[0]
        category_count = data[data[feature] == category].shape[0]
        try:
            ratio = 100 * (outcome_count / category_count)
        except ZeroDivisionError:
            ratio = 0
        category_to_ratio[category] = ratio

    sorted_ratios = dict(sorted(category_to_ratio.items(), key=lambda item: item[1], reverse=True))
    categories = list(sorted_ratios.keys())
    ratios = list(sorted_ratios.values())

    plt.figure(figsize=(10, 6))
    sns.barplot(x=categories, y=ratios, palette=selected_colormap)
    plt.xlabel(feature)
    plt.ylabel(f'Percentage of {target_outcome} in {target}')
    plt.title(f'{target} Rate by {feature}')
    plt.ylim(min(ratios) - 2, max(ratios) + 1)
    plt.show()

    # Printing the ratios
    for category, ratio in sorted_ratios.items():
        print(f'{target} Rate for category {category.upper()} in feature {feature.upper()} is {ratio:.2f}%')
```

```python
analyze_categorical_relationship('Warehouse_Block', 'Delivered_On_Time', data)
analyze_categorical_relationship('Shipment_Mode', 'Delivered_On_Time', data)
analyze_categorical_relationship('Product_Importance', 'Delivered_On_Time', data)
analyze_categorical_relationship('Customer_Gender', 'Delivered_On_Time', data)
```

```python
# Function to analyze numerical features w.r.t target
def explore_numerical_feature(feature_name, target, data):
    print(f"Exploring {str(feature_name).upper()}........")
    print(f"Mean of {feature_name}: {data[feature_name].mean()}")
    print(f"Median of {feature_name}: {data[feature_name].median()}")

    # Correlation with target
    temp = data[[feature_name, target]]
    corr = temp.corr().iloc[0, 1]
    print(f"Correlation with the target feature ({target}): {corr}")

    # Plotting
    plt.figure(figsize=(17, 5))
    plt.subplot(121)
    sns.kdeplot(data=data, x=feature_name, hue=target)
    plt.title(f'Distribution of {feature_name} by {target}')

    plt.subplot(122)
    sns.boxplot(x=target, y=feature_name, data=data)
    plt.title(f'Boxplot of {feature_name} by {target}')

    plt.show()
```

```python
explore_numerical_feature('Customer_Care_Calls', 'Delivered_On_Time', data)
explore_numerical_feature('Customer_Rating', 'Delivered_On_Time', data)
explore_numerical_feature('Product_Cost', 'Delivered_On_Time', data)
explore_numerical_feature('Prior_Purchases', 'Delivered_On_Time', data)
explore_numerical_feature('Discount_Offered', 'Delivered_On_Time', data)
explore_numerical_feature('Product_Weight', 'Delivered_On_Time', data)
```

# Feature Generation

```python
# Discount to Cost Ratio
data['Discount_to_Cost_Ratio'] = data['Discount_Offered'] / data['Product_Cost']
```

```python
# Total Customer Interactions
data['Total_Customer_Interactions'] = data['Customer_Care_Calls'] + data['Prior_Purchases']

# Customer Loyalty
data['Customer_Loyalty'] = data['Prior_Purchases'].apply(lambda x: 1 if x >= 5 else 0)
```

# Data Preparation

```python
# Drop Customer_ID
data.drop(columns=['Customer_ID'], inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder, MinMaxScaler

# Encode categorical features
label_encoders = {}     # Dictionary to store encoders

for feature in ['Warehouse_Block', 'Shipment_Mode', 'Product_Importance', 'Customer_Gender']:
    le = LabelEncoder()
    data[feature] = le.fit_transform(data[feature])
    label_encoders[feature] = le   # Storing trained encoder
```

```python
# Normalizing numerical features
numerical_features = ['Customer_Care_Calls', 'Customer_Rating', 'Product_Cost', 'Prior_Purchases', 'Discount_Offered', 'Product_Weight']
scaler = MinMaxScaler()
data[numerical_features] = scaler.fit_transform(data[numerical_features])
```

```python
# Separating features and target
X = data.drop(columns=['Delivered_On_Time'])
y = data['Delivered_On_Time']
```

```python
visualize_categorical(y, "Distribution of Target")
```

```python
from imblearn.over_sampling import SMOTE

# Balancing data using SMOTE
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)
```

```python
visualize_categorical(y_resampled, "Distribution of Target")
```

```python
# Preparing dataframe of balanced data
resampled_data = pd.DataFrame(X_resampled, columns=X.columns)
resampled_data['Delivered_On_Time'] = y_resampled
resampled_data.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.feature_selection import SelectKBest, f_classif

# Calculating correlation with target
correlations = resampled_data.corr()['Delivered_On_Time'].drop('Delivered_On_Time')
correlation_sorted = correlations.sort_values(ascending=False)

print(correlation_sorted)
# Plot correlations with target
plt.figure(figsize=(10, 6))
correlation_sorted.plot(kind='bar', title='Correlation with Target')
plt.xlabel('Features')
plt.ylabel('Correlation with Target')
plt.show()
```

```python
# SelectKBest to select the top 10 features
selector = SelectKBest(score_func=f_classif, k=10)
X_resampled = pd.DataFrame(X_resampled, columns=X.columns)  # Ensure X_resampled has correct column names
selector.fit(X_resampled, y_resampled)
selected_features = selector.get_support(indices=True)
selected_feature_names = X_resampled.columns[selected_features]
scores = selector.scores_[selected_features]

# Ploting feature importance
plt.figure(figsize=(10, 6))
sns.barplot(x=scores, y=selected_feature_names)
plt.title('Top 10 Features by Importance Score')
plt.xlabel('Importance Score')
plt.ylabel('Feature')
plt.show()

# Displaying the selected features and their scores
for feature, score in zip(selected_feature_names, scores):
    print(f'{feature}: {score}')
```

```python
# Selecting top features
X_sel = X_resampled[selected_feature_names]
y_sel = y_resampled
```

```python
# Split the data into train and test sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_sel, y_sel, test_size=0.2, random_state=42)
```

# Modeling

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import BaggingClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report, ConfusionMatrixDisplay
import time


# Models
models = {
    'DecisionTreeClassifier': DecisionTreeClassifier(),
    'MLPClassifier': MLPClassifier(),
    'GradientBoostingClassifier': GradientBoostingClassifier(),
    'BaggingClassifier': BaggingClassifier()
}
```

```python
# Function to evaluate models
def evaluate_model(model, X_train, y_train, X_test, y_test):
    start_time = time.time()

    # Fitting model to the Training set
    model.fit(X_train, y_train)

    # Predict values
    y_pred = model.predict(X_test)

    # Evaluate metrics
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred, average='weighted')
    recall = recall_score(y_test, y_pred, average='weighted')
    f1_sco = f1_score(y_test, y_pred, average='weighted')
    confusion_mat = confusion_matrix(y_test, y_pred)
    report = classification_report(y_test, y_pred)

    print("+", "="*100, "+")
    print(f"\t\t\tFor {model.__class__.__name__} The Performance result is:")
    print("+", "="*100, "+")
    print('Accuracy : ', accuracy)
    print("-"*50)
    print('F1 : ', f1_sco)
    print("-"*50)
    print('Recall : ', recall)
    print("-"*50)
    print('Precision : ', precision)
    print("-"*50)
    print('Confusion Matrix....\n', confusion_mat)
    print("-"*50)
    print('Classification Report....\n', report)
    print("-"*50)
    print('Plotting Confusion Matrix...\n')
    ConfusionMatrixDisplay(confusion_mat).plot()
    plt.show()

    print("\t\t\t-----------------------------------------------------------")
    print(f"\t\t\t Time for detection ({model.__class__.__name__}) : {round((time.time() - start_time), 3)} seconds...")
    print("\t\t\t-----------------------------------------------------------")

    return {
        'model': model.__class__.__name__,
        'accuracy': accuracy,
        'precision': precision,
        'recall': recall,
        'f1_score': f1_sco,
        'confusion_matrix': confusion_mat
    }, model
```

```python
# Evaluate all models
results = []
for model_name, model in models.items():
    result, model = evaluate_model(model, X_train, y_train, X_test, y_test)
    results.append(result)
```

```python
# DataFrame to display results
results_df = pd.DataFrame(results)
results_df
```

# Hyperparameter Tuning

```python
from sklearn.model_selection import GridSearchCV

gbc = GradientBoostingClassifier()

# Parameter grid
param_grid = {
    'n_estimators': [50, 100, 150],
    'learning_rate': [0.01, 0.1, 0.2],
    'max_depth': [3, 4, 5],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'subsample': [0.8, 0.9, 1.0]
}

# GridSearchCV
grid_search = GridSearchCV(estimator=gbc, param_grid=param_grid, 
                           scoring='accuracy', cv=2, n_jobs=-1, verbose=2)
grid_search.fit(X_train, y_train)
best_params = grid_search.best_params_

# Printing the best parameters and the best score
print(f"Best parameters found: {grid_search.best_params_}")
print(f"Best accuracy score: {grid_search.best_score_}")
```

```python
# Training and Evaluating Final Model
final_model = GradientBoostingClassifier(**best_params)
results, final_model = evaluate_model(final_model, X_train, y_train, X_test, y_test)
```

```python
# Feature importance from the final model
feature_importances = final_model.feature_importances_
feature_importances_df = pd.DataFrame({
    'Feature': X_train.columns,
    'Importance': feature_importances
})

# Sorting the DataFrame by importance
feature_importances_df = feature_importances_df.sort_values(by='Importance', ascending=False)

# Plot the feature importances
plt.figure(figsize=(12, 8))
sns.barplot(x='Importance', y='Feature', data=feature_importances_df)
plt.title('Feature Importance from Random Forest Model')
plt.xlabel('Importance')
plt.ylabel('Feature')
plt.show()
```
